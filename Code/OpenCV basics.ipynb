```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('watch.jpg',cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.plot([200,300,400],[100,200,300],'c', linewidth=5)
plt.show()
```

```python
import numpy as np
import cv2

cap = cv2.VideoCapture(0)  #loading the data from the webcam...
 
while(True):
    ret, frame = cap.read() # ret is a boolean variable
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
 
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

img = cv2.imread('watch.jpg',cv2.IMREAD_COLOR)
```

```python
cv2.line(img,(0,0),(150,150),(255,255,255),10) # parameters : start coordinates,end-points, color(b,g,r), thickness

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# for a rectangle
cv2.rectangle(img,(15,25),(200,150),(0,0,255),15) # parameters : start coordinates,end-points, color(b,g,r), thickness

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# for a circle
cv2.circle(img,(100,63), 55, (0,255,0), -1) # parameters: start coordinates,end-points, color(b,g,r), -1 to get a filled circle 

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# for a polygon
pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
# OpenCV documentation had this code, which reshapes the array to a 1 x 2. I did not 
# find this necessary, but you may:
#pts = pts.reshape((-1,1,2))
cv2.polylines(img, [pts], True, (0,255,255), 3)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

img = cv2.imread('watch.jpg',cv2.IMREAD_COLOR)
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,'My Watch!',(0,130), font, 1, (200,255,155), 2, cv2.LINE_AA)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
